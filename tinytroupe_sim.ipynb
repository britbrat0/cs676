```python
# ===============================
# 0️⃣ Setup & Imports
# ===============================

import os

# ---- Set your OpenAI API key here ----
# Replace with your actual key
os.environ["OPENAI_API_KEY"] = ""

# TinyTroupe imports
import tinytroupe
from tinytroupe.agent.tiny_person import TinyPerson
from tinytroupe.environment.tiny_world import TinyWorld
from tinytroupe.extraction.results_extractor import ResultsExtractor
```

```python
# Clear previous agents and worlds
TinyPerson.all_agents.clear()
TinyWorld.all_environments.clear()
```

```python
# Context 1: High stress situation
world_stress = TinyWorld("StressfulSituation", [dave, eve, carol])
world_stress.make_everyone_accessible()

# Context 2: Work environment
world_work = TinyWorld("WorkMeeting", [sandra, alice, bob])
world_work.make_everyone_accessible()

# Context 3: Data science conversation
world_ds = TinyWorld("DSconvo", [brenna, arnold, mew])
world_ds.make_everyone_accessible()
```

```python
# High stress situation
dave.listen("I think there has been a security breach.")

# Work environment
sandra.listen("Can someone fill me in on what I missed in the meeting?")

# DS convo
brenna.listen("How can we use data science to classify flowers?")
```

```python
world_ds.run(9)
```

```python
extractor = ResultsExtractor()

summary_stress = extractor.extract_results_from_world(
    world_stress,
    extraction_objective="Summarize conversation from each person's point of view",
    verbose=False
)

summary_work = extractor.extract_results_from_world(
    world_work,
    extraction_objective="Summarize conversation from each person's point of view",
    verbose=False
)


summary_ds = extractor.extract_results_from_world(
    world_ds,
    extraction_objective="Summarize conversation from each person's point of view",
    verbose=False
)

print("=== Stressful Situtation Conversation Summary ===")
print(summary_stress)

print("\n=== Work Meeting Summary ===")
print(summary_work)


print("\n=== DS Convo Summary ===")
print(summary_ds)
```
