```python
# ==========================
# Simulate Focus Group
# ==========================

features_to_test = [
    "Smart Daily Summary – AI-powered recap of messages, meetings, and to-do progress",
    "Contextual Recommendation Feed – Personalized content and app tips based on usage patterns",
    "Accessibility Co-Pilot – Voice assistant that adjusts layout, color contrast, and text size dynamically"
]

conversation_log = []

for feature in features_to_test:
    initial_prompt = f"Let's discuss {feature}. What are your thoughts on this idea?"

    # Send initial message to all agents
    for agent in world_focus_group.agents:
        agent.listen(initial_prompt)

    # Run the world for a few turns
    world_focus_group.run(3)

    # Collect messages per agent
    feature_entry = {"feature": feature, "responses": {}}
    for agent in world_focus_group.agents:
        messages = getattr(agent, "current_messages", [])
        if messages:
            # Convert any dict messages to string
            text_messages = []
            for msg in messages:
                if isinstance(msg, dict):
                    text_messages.append(msg.get("text", str(msg)))
                else:
                    text_messages.append(str(msg))
            feature_entry["responses"][agent.name] = text_messages
        else:
            feature_entry["responses"][agent.name] = ["_No interactions recorded for this persona._"]

    conversation_log.append(feature_entry)
```

```python
extractor = ResultsExtractor()

summary = extractor.extract_results_from_world(
    world_focus_group,
    extraction_objective="Summarize conversation from each person's point of view",
    verbose=False
)

print("\n=== Focus Group Summary ===")
print(summary)
```
